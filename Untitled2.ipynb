### Visualizing road traffic data 
#### Uncovering patterns in the traffic 

```python
import pandas as pd
```

```python
df1 = pd.read_csv("RilevazioniPerPostazionev2.csv", sep = ";" )
df2 = pd.read_csv("RilevazioniPerPostazionev1.csv", sep = ";" )
data = pd.concat([df1, df2], ignore_index=True)
data.shape
```

```python
data.isna().sum()
```

```python
data.head()
```

```python
data.columns
```

```python
cols = ['YearMonth', 'SensorNr', 'DayMonthYear', 'DayNr', 'RoadNameNumber','TotalVehicles', 
        'NonClassVeh', 'LightWeightVehicles', 'HeavyVehicles', 'DayTimeTraff', 'NightTimeTraff',
       'WeekdaysTraff', 'WeekendsTraffic']
```

```python
data.columns = cols
```

```python
data
```

```python
gbrn = data.groupby('RoadNameNumber')
for i, k in gbrn: 
    print(i)
data2 = gbrn.get_group('NSA 313 complanare sud alla A 14 nel tratto tra San Lazzaro e Idice')
```

```python
def fix_date(x): 
    #14/08/2025
    x = x.replace('/', '-')
    return x
    
data2['DayMonthYear'] = data2['DayMonthYear'].apply(lambda x : fix_date(x))
```

```python
data2viz = data2[['DayMonthYear', 'HeavyVehicles']].sort_values(by='DayMonthYear', ascending=True).set_index('DayMonthYear')
data2viz
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (20, 5))
plt.plot(data2viz)
plt.xticks(rotation = 45)
plt.show()
```
